```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wine_data=pd.read_csv('winequalityred.csv')
wine_data
```

```python
X = wine_data.drop("quality",axis=1)
y= wine_data["quality"]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
df_exp=X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax_scale = MinMaxScaler()

minmax_scale.fit((df_exp))

df_exp_minmax = minmax_scale.transform(df_exp)
print(type(df_exp_minmax))
df_exp_minmax_new = pd.DataFrame(df_exp_minmax,columns=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol'])
df_exp_minmax_new
```

```python
X_test_minmax = minmax_scale.transform(X_test)
print(type(X_test_minmax))
df_test_minmax_new = pd.DataFrame(X_test_minmax,columns=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol'])
df_test_minmax_new
```

```python
ax = df_exp.plot.scatter(x="fixed acidity", y="density", marker = "P",s=80, label='BREFORE SCALING',color='red')
df_exp_minmax_new.plot.scatter(x="fixed acidity", y="density",marker = "x",s=60,label='AFTER SCALING',
                               color='blue', ax = ax)
plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```
