```python
import pandas as pd
import numpy as np
```

```python
n=['class','age','menopause','tumor_size','inv_nodes','node_caps','deg_malig','breast','breast_quad','irradiate']
df = pd.read_csv('breast-cancer.data',sep = ',', header =None, names = n)
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in range(10):
    df.iloc[:,i] = le.fit_transform(df.iloc[:,i])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df);
```

```python
sns.heatmap(df.corr(),cmap='RdYlGn');
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df , hue = 'class');
```

```python
df.corr().style.background_gradient(cmap='coolwarm')
```

```python
for i in n:
    sns.kdeplot(df[i],shade=True)
    plt.title(f"Distribution of {i}")
    plt.xlabel(i)
    plt.ylabel('Density')
    plt.show()
```

```python
y = df['class']
x = df.iloc[:,1:]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5)
```

# NaiveBayes(0 hyperparameters, 1model)

```python
#NaiveBayes
print('\n--------------------------------------------------------------------\n')

gn = GaussianNB()
gn.fit(x_train,y_train)
y_pred1 = gn.predict(x_test)
print("Accuracy of NaiveBayes :",accuracy_score(y_test,y_pred1))
print("Classification report for NaiveBayes :\n",classification_report(y_test,y_pred1))

print('\n--------------------------------------------------------------------\n')

```

# SVM(0 hyperparameters, 1model)

```python
#SVM
print('\n--------------------------------------------------------------------\n')

svm = SVC()
svm.fit(x_train,y_train)
y_pred2 = svm.predict(x_test)
print("Accuracy of SVM :",accuracy_score(y_test,y_pred2))
print("Classification report for SVM :\n",classification_report(y_test,y_pred2))
print('\n--------------------------------------------------------------------\n')
```

# K-NN(1 hyperparameters, 4 models)

```python
#KNN
print('\n---------------------------------(K=3)-----------------------------------\n')

knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_3.fit(x_train,y_train)
y_pred = knn_3.predict(x_test)
print("Accuracy of KNN :",accuracy_score(y_test,y_pred))
print("Classification report for KNN\n",classification_report(y_test,y_pred))

print('\n---------------------------------(K=5)-----------------------------------\n')

knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(x_train,y_train)
y_pred = knn_5.predict(x_test)
print("Accuracy of KNN :",accuracy_score(y_test,y_pred))
print("Classification report for KNN\n",classification_report(y_test,y_pred))

print('\n---------------------------------(K=7)-----------------------------------\n')

knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_7.fit(x_train,y_train)
y_pred = knn_7.predict(x_test)
print("Accuracy of KNN :",accuracy_score(y_test,y_pred))
print("Classification report for KNN\n",classification_report(y_test,y_pred))

print('\n---------------------------------(K=13)-----------------------------------\n')

knn_13 = KNeighborsClassifier(n_neighbors=13)
knn_13.fit(x_train,y_train)
y_pred = knn_13.predict(x_test)
print("Accuracy of KNN :",accuracy_score(y_test,y_pred))
print("Classification report for KNN\n",classification_report(y_test,y_pred))

print('\n--------------------------------------------------------------------\n')
```

```python
total_Scores=list()
```

```python
for n in range(3,6):
    cv = KFold(n_splits=n, random_state=1, shuffle=True)
    scores = cross_val_score(knn_13, x, y, scoring='accuracy',cv=cv)
    total_Scores.append(scores)
    #view mean accuracy
    print(np.mean(scores))

```

```python
for n in range(3,6):
    cv = KFold(n_splits=n, random_state=1, shuffle=True)
    scores = cross_val_score(svm, x, y, scoring='accuracy',cv=cv)
    total_Scores.append(scores)
    #view mean accuracy
    print(np.mean(scores))

```

```python
for n in range(3,6):
    n_splits=n
    cv = KFold(n_splits, random_state=1, shuffle=True)
    scores = cross_val_score(gn, x, y, scoring='accuracy',cv=cv)
    total_Scores.append(scores)
    #view mean accuracy
    print(np.mean(scores))

```

```python
# compute final score of L 

final_perf = sum(scores) / n_splits 

print('Final performance of L:', final_perf)
```
